Unsupervised preprocessing and visualization

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Datensatz laden
food_df = pd.read_csv("food.csv")

# Standardisierung
X_food = food_df.drop(columns=["Unnamed: 0"])
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_food)

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot PC1/PC2
plt.figure(figsize=(10, 7))
plt.scatter(X_pca[:, 0], X_pca[:, 1])

# Country-Labels hinzufgen
for i, country in enumerate(food_df["Unnamed: 0"]):
    plt.text(X_pca[i, 0] + 0.1, X_pca[i, 1], country, fontsize=9)

plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')
plt.title("PCA Score Plot (PC1 vs. PC2)")
plt.grid(True)
plt.show()
```
